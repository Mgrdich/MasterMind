```python
from main import MasterMindBinary
import matplotlib.pyplot as plt
import numpy as np

plt.rcdefaults()
```

```python
masterMind = MasterMindBinary(auto_play=True, show_computer_guess=True)
masterMind.play_game()
masterMind.print_results()
```

```python
history, bot_guesses, whole_combination = masterMind.get_results_of_iterations()
print("History", history)
print("Bot guess", bot_guesses)
print("whole Combination", whole_combination)
```

```python
for specific_history in history:
    y_values = []

    for i in whole_combination:
        value = i in specific_history
        y_values.append(int(value))


    plt.scatter(whole_combination, y_values)
    #set parameters for tick labels
    plt.tick_params(axis='x', which='major', labelsize=6)
    plt.show()
```
